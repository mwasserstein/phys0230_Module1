1. Plot $K_r$ and $K_N$ as functions of $v/c$ on the same plot.

```python
x_values = np.linspace(0, 2, 200)
K_n = (1 / 2) * x_values ** 2
K_r = 1 / np.sqrt(1 - x_values ** 2) - 1

plt.plot(x_values, K_n, label=r'$K_N$ / $E_0$')
plt.plot(x_values, K_r, label=r'$K_R$ / $E_0$')
plt.xlabel('v / c', fontsize=16)
plt.ylabel('Ratio of Kinetic Energy to Rest Energy')
plt.legend()
```

$K_r$ and $K_N$ agree when $v$/$c$ is small ($v$ << $c$), and they disagree as $v$/$c$ approaches 1.

My plot does give an indication why a massive particle must always travel at speeds less than $c$. The orange line in my plot above is asymptotically bounded at $v$/$c$ = 1. That means that $v$ cannot be larger than $c$ when we consider relativity. Kinetic energy gets infinitely large as v approaches $c$, and it keeps rising. Thus, $v$ can never be larger than $c$.

<font color=red>2.*</font> $K_N$ is an *approximation* to $K_r$ that is accurate to 2nd order in $\beta \equiv v/c$.

* Find the next lowest order correction $\delta K$ to $K_N$ such that $K_r \simeq K_N + \delta K$.
* Plot the absolute error ($ | K_N - K_r |$) error in $K_N$ and $\delta K$ vs. $v/c$. Explain what you plots shows.

$$ \delta K = -\frac{3}{8} m \frac{v^4}{c^2} $$

Therefore,

$$ K_r \simeq \frac{1}{2}m v^2 -\frac{3}{8} m \frac{v^4}{c^2} $$

(See my attached supplemental work)

```python
x_values = np.linspace(0, 1, 200)  # beta
K_n = (1 / 2) * x_values ** 2
K_r = 1 / np.sqrt(1 - x_values ** 2) - 1
delta_K = np.abs(- (3 / 8) * x_values ** 4)

K_n_minus_K_r = np.abs(K_n - K_r)

plt.plot(x_values, K_n_minus_K_r, label=r'$|K_N - K_R|$')
plt.plot(x_values, delta_K, label=r'$\delta_K$')
plt.xlabel(r'$\frac{v}{c}$', fontsize=20)
plt.yscale('log')
plt.title(r'Absolute Error in $K_N$ and $\delta_K$')
plt.legend()
```

We know that $K_r \simeq K_N + \delta K$, which implies that $\delta K \simeq K_r - K_N$. Therefore, $|\delta K|$ should equal $|K_r - K_N|$ for most values of $\beta$ (or $\frac{v}{c}$). This holds true for small values of $v$, meaning that the Newtonian approximation is a good one when $v$ is small. However, as $v$ approaches $c$, $|K_r - K_N|$ and $|\delta K|$ diverge from eachother (as my plot shows) meaning that $K_N$ is a poor approximation for $K_R$ for values of $v$ near $c$.

3. Consider two cases:
    * A force is applied in a direction *perpendicular* to the velocity $\vec{F} \perp \vec{v}$.
    * A force is applied in the same direction as the velocity $\vec{F} \parallel \vec{v}$.
   
   For each case plot the ratio of the magnitude of the relativistic acceleration to the magnitude of the Newtonian acceleration as a function of $\beta$.

```python
betas = np.linspace(0, 1, 200) # assume 1 is max value of beta since v cannot be larger than c

ratio_perpen = np.sqrt(1 - betas ** 2)
ratio_parra = np.power(1 - betas ** 2, 3 / 2)


plt.plot(betas, ratio_perpen, label=r'$\vec{F}$ and $\beta$ are perpendicular')
plt.plot(betas, ratio_parra, label=r'$\vec{F}$ and $\beta$ are parallel')
plt.xlabel(r'$\beta$', fontsize=20)
plt.ylabel(r'$\frac{\vec{a}_r}{\vec{a}_N}$', fontsize=20)
plt.legend()
```

It is easier to accelerate a particle along the direction its moving. My plots above show that $a_r$ / $a_N$ slopes down faster when $\vec{F}$ and $\vec{\beta}$ are parallel. In other words, except when $|\beta|$ is 0 or 1, $a_r$ / $a_N$ is smaller when $\vec{F}$ and $\vec{\beta}$ are parallel, meaning that a particle can accelerate more easily than when $\vec{F}$ and $\vec{\beta}$ are perpendicular. Therefore, it would be easier to accelerate a particle at a given speed ($\beta$) if $\vec{F}$ and $\vec{\beta}$ are parallel.

In both of my plots above, when v = c ($\beta$ = 1), the ratio of relativistic acceleration to Newtonian acceleration is zero, meaning that relativistic acceleration is zero. Therefore, a massive particle can never travel at the speed of light, since its accelration would have to be be zero. If its acceleration were zero, it would not be able to change its motion, meaning that it cannot travel at the speed of light.

4.* Use the relativistic expression for momentum, to find the speed of a relativistic particle accelerated from rest as a function of time. Notice that your expression for  /  is a funciton of the dimensionless quantity  / .

$$ \frac{v_r}{c} = \frac{\tau}{\sqrt{1+\tau^2}}, $$

where 

$$ \tau = Ft/mc $$ 

See my supplemental work for how I found this result.

Plot the speed of the particle vs. time. For the y-axis plot the ratio  $v_r/c$  and for the x-axis plot the dimensionless quanity  / .

```python
v_r_over_c = np.linspace(0, 1, 200)
tau = np.linspace(0, 10, 200)
x_values = tau / np.sqrt(1 + tau ** 2)
plt.plot(x_values, v_r_over_c)
plt.xlabel(r'$\tau$', fontsize=20)
plt.ylabel(r'$v_r/c$', fontsize=20)
```

* What happens to the speed of the particle if a constant force is applied indefinitely?

If a constant force is applied indefinitely, the speed of the particle approaches the speed of light ($c$). This is apparent in my plot because as $\tau$ increases, $v_r/c$ approaches 1. However, the speed of the particle can never increase above the speed of light, regardless of how large $\tau$ gets.

* Is any part of your plot consistent with Newtonian physics? If so, explain.

In Newtonian physics, $$ v_N = \frac{F}{m} t, $$ meaning that velocity $v_N$ is proportional to $t$, and a plot of $v_N$ as a function of time should be linear.
